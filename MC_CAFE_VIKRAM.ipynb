```python

menu = {
    "Coffee": {
        "Espresso": 1.50,
        "Americano": 2.00,
        "Caffe latte": 2.25,
        "Mocha": 2.25,
        "Cappucino": 2.50,
        "Macchiato": 2.50
    },
    "Tea": {
        "White tea": 1.50,
        "Green tea": 2.00,
        "Oolong tea": 2.25,
        "Black tea": 2.25,
        "Earl Grey": 2.50,
        "Matcha": 2.50
    },
    "Pastries": {
        "Croissant": 5.50,
        "Cinnamon roll": 4.50,
        "Banana bread": 4.25,
        "Pumpkin bread": 3.75,
        "Sticky bun": 4.25,
        "Apple turnover": 3.75
    }
}

def display_menu():
    for category, items in menu.items():
        print(f"\n{category}:")
        for item, price in items.items():
            print(f"{item} - ${price:.2f}")

def get_order():
    order = {}
    while True:
        item = input("\nEnter the item you would like to order: ").title()
        if any(item in menu[category] for category in menu):
            quantity = int(input("Enter the quantity: "))
            order[item] = order.get(item, 0) + quantity
        else:
            print("Sorry, that item is not on the menu.")
        
        more_items = input("Would you like to order another item? (yes/no): ").lower()
        if more_items != "yes":
            break
    return order

def print_receipt(order):
    subtotal = 0
    print("\nYour order:")
    for item, quantity in order.items():
        for category in menu:
            if item in menu[category]:
                price = menu[category][item]
                total_price = price * quantity
                print(f"{item} x{quantity} - ${total_price:.2f}")
                subtotal += total_price
                break

    sales_tax = subtotal * 0.06
    grand_total = subtotal + sales_tax
    print(f"\nSubtotal: ${subtotal:.2f}")
    print(f"Sales Tax: ${sales_tax:.2f}")
    print(f"Grand Total: ${grand_total:.2f}")

def main():
    print("Welcome to MC Cafe!")
    if input("Would you like to see the menu? (yes/no): ").lower() == "yes":
        display_menu()

    if input("Would you like to place an order? (yes/no): ").lower() == "yes":
        order = get_order()
        print_receipt(order)

    print("Thank you for visiting MC Cafe!")

if __name__ == "__main__":
    main()
```
